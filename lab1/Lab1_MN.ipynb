<h3> Tipuri de date și variabile</h3>

Python oferă **tipuri de date** numerice, booleene, șiruri (string, liste etc), dicționare, fișiere, clase, instanțe și excepții.

Din punct de vedere al tipării Python folosește tipuri pentru obiecte, <i>însă la definirea variabilelor nu trebuie precizat tipul acestora</i>. Constrângerile de tip sunt verificate la execuție (late binding), astfel încât pot apărea erori și excepții generate de folosirea unui tip necorespunzător în atribuiri și excepții.

Python asociază numele unei variabile cu un obiect, care poate fi număr, șir de caractere sau ceva mai complex. Când este folosită o variabilă, tipul acesteia este tipul obiectului cu care este asociată. Este greșită folosirea într-o expresie a unei variabile care nu a fost asociată cu un obiect.


```python
s = "Hi"   # s va fi de tip String
k = 5     # k va fi de tip Integer
print(s,k)

#[cerinta]: Utilizati functia type() pentru a afla tipul de date al variabilei s
```

În Python se poate face atribuirea de valori unor mai multe variabile simultan:

```python
x,y = 2,3
print(x,y)
```

Datorită faptului că partea dreaptă a unei expresii este evaluată înainte de a se face atribuirea, valorile a două variabile pot fi foarte ușor interschimbate, fară a avea nevoie de o a treia variabilă:

```python
x, y = y, x
print(x,y)
```

<h3>Funcții</h3>
Funcțiile sunt definite folosind cuvântul cheie <b style="color:##228B22">def</b> urmat de numele functiei, o listă de argumente și apoi de ”:”.
<table>
 <li>lista de argumente conține doar numele acestora, tipul lor fiind cel trimis la apelul funcției.</li>
 <li>se pot specifica valori implicite pentru parametrii unei funcții, iar dacă la apelare nu se include acel parametru este luată valoarea implicită</li>
 <li>dacă nu se include <b style="color:##228B22">return</b> sau se execută return fară argumente, atunci funcția va întoarce None.</li>
</table>

```python
#Exemplu de definire a unei functii
def fractie(x, y=1):   # y are valoare implicita
    if (y==0):
        return         # va întoarce None
    else:
        return x/y

print(fractie(10))
```

<h3>Includerea de cod extern</h3>
În Python se pot include funcții, variabile și clase definite în alte fișiere. Un astfel de fișier, ce poate fi inclus, poartă denumirea de modul. Acest lucru se realizează cu ajutorul instrucțiunii <b style="color:##228B22">import</b>.

```python
import numpy
print(numpy.sqrt(9)) # ex de utilizare a functiei sqrt() din cadrul modulului numpy
```

Atunci când includem modulul putem să îl folosim cu un alt nume, eventual, pentru a evita coliziuni de denumiri:

```python
import numpy as np        #ex de utilizare a unui modul sub alt nume
```

Un alt mod de a include este folosind <b style="color:##228B22">from </b> <i>nume_modul </i> <b style="color:##228B22">import</b> <i>lista_constructii</i>, în felul acesta vom avea acces doar la clasele/metodele din modul precizate în <i>lista_constructii</i>. Această modalitate este indicată atunci când un modul are un număr foarte mare de clase. Un alt avantaj al acestei metode este că nu se mai folosește numele modulului atunci când se utilizează această construcție, ca în exemplul de mai jos:

```python
from numpy import linalg as la

a = np.array([ 4, 3, 0, 0]) #declararea unui vector
norma = la.norm(a)
print(norma)
```
